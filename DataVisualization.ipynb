```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
x = np.arange(0,10)
y=np.arange(10,20)
```

```python
plt.scatter(x,y, c ="g", marker="*")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Sctter plot using matplotlib")
plt.show()
```

```python
plt.plot(x,y,"r")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("plot line graph")
plt.show()
```

```python
X=["a","b","c"]
Y=[50,60,70]
```

```python
plt.bar(X,Y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("plot bar graph")
plt.show()
```

```python
df=[10,20,30,40,50,60,70,200,-100]
plt.boxplot(df)
plt.show
```

```python
x=[10,60,20,20]
labels=["a","b","c","d"]
colors=["r","b","p","w"]
explode=[0,0.1,0,0]
```

```python
plt.pie(x,explode=explode,labels=labels,autopct="%1.3f%%",shadow=True)
plt.title("pie chart")
plt.show()


```
